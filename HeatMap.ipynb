```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

```python
GlobalWarming_df = pd.read_csv(r"D:\New folder\responsible_global_warming.csv")
GlobalWarming_df
```

```python
GlobalWarming_df = GlobalWarming_df.drop(columns = ["Country Code","Indicator Name", "Indicator Code"], axis = 1).set_index("Country Name")
GlobalWarming_df.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(GlobalWarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(GlobalWarming_df, cmap = "coolwarm", vmin = 0, vmax = 21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(GlobalWarming_df, cmap = "coolwarm", vmin = 0, vmax = 21, annot = True)
plt.show()
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {"orientation" : "horizontal", "shrink" : 1, "extend" : 1, "ticks" : np.arange(0,22),}
sns.heatmap(GlobalWarming_df, cbar= False, xticklabels = False, yticklabels = False,)
plt.show()
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {"orientation" : "horizontal", "shrink" : 1, "extend" : 1, "ticks" : np.arange(0,22),}
sns.heatmap(GlobalWarming_df, cbar_kws = cbar_kws)
plt.show()
```
