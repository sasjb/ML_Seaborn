Seaborn Histogram Plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.histplot(tips_df["tip"])
plt.show()
```

```python
sns.distplot(tips_df["total_bill"])
plt.show()
```

```python
sns.distplot(tips_df["total_bill"], bins = 100)
plt.show()
```

```python
sns.distplot(tips_df["total_bill"], label = "Total Bill")
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"], bins = bins)
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"], bins = bins, hist_kws = {'color': "y", 'edgecolor': "r", 'linewidth': 5,})
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"], bins = bins, hist_kws = {'color': "y", 'edgecolor': "r", 'linewidth': 5,'alpha':0.8})
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"], bins = bins, hist_kws = {'color': "y", 'edgecolor': "r", 'linewidth': 5,'alpha':0.8},
            kde_kws = {'color': "m", 'linewidth': 5,'alpha':0.8})
plt.xticks(bins)
plt.title("Histogram of Total Bill", fontsize = 20)
plt.legend()
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"], bins = bins, hist_kws = {'color': "y", 'edgecolor': "r", 'linewidth': 5,'alpha':0.8},
             kde=False,fit = norm, fit_kws = {'color': "m", 'linewidth': 5,'alpha':0.8},
            )
plt.xticks(bins)
plt.title("Histogram of Total Bill", fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.distplot(tips_df["total_bill"], label = "total_bill")
sns.distplot(tips_df["tip"], label = "tip")
sns.distplot(tips_df["size"], label = "size")
```
