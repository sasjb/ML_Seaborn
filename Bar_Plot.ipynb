Seaborn Bar Plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
plt.show()
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex" , data = tips_df , alpha = 0.9)
plt.show()
```

```python
order = ['Sat','Sun','Thur','Fri']
hue_order = ['Female','Male']
sns.barplot(x = "day", y = "total_bill", hue = "sex" , data = tips_df , order = order, hue_order = hue_order)
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex" , data = tips_df , order = order, hue_order = hue_order, estimator = np.max)
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex" , data = tips_df , order = order, hue_order = hue_order, estimator = np.max,
           ci = 50 , palette = 'hot')
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "sex" , data = tips_df , palette = 'hot', saturation = 0.5)
plt.show()
```

```python
kwargs = {'alpha' : 0.9, 'linestyle': ':', 'linewidth' : 10, 'edgecolor' : 'k'}
sns.barplot(x = "day", y = "total_bill", hue = "sex" , data = tips_df , **kwargs)
plt.show()
```

```python
sns.set()
kwargs = {'alpha' : 0.9, 'linestyle': ':', 'linewidth' : 10, 'edgecolor' : 'k'}
sns.barplot(x = "day", y = "total_bill",  data = tips_df , **kwargs)
plt.show()
```

```python
sns.set()
kwargs = {'alpha' : 0.9, 'linestyle': '-.', 'linewidth' : 2, 'edgecolor' : 'g'}
ax = sns.barplot(x = "day", y = "total_bill",  data = tips_df , **kwargs)
ax.set(title = "Bar Plot of Tip DataFrame",
      xlabel = "Day",
      ylabel = "Total Bill")
plt.show()
```

```python
sns.set()
kwargs = {'alpha' : 1, 'linestyle': '-.', 'linewidth' : 2, 'edgecolor' : 'k'}
ax = sns.barplot(x = "day", y = "total_bill",  data = tips_df , **kwargs)
plt.title("Bar Plot of Tip DataFrame", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Total Bill",fontsize = 15)
plt.show()
```
